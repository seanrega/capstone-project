# Factor analysis: Chi-square

```python
from sklearn.feature_selection import SelectKBest 
from sklearn.feature_selection import chi2 
  

  
# Create features and target 
X = df_nsdropnullx
y = df_nsdropnull['income_poor']
  
# Convert to categorical data by converting data to integers 
X = X.astype(int) 

  
# Two features with highest chi-squared statistics are selected 
chi2_features = SelectKBest(chi2, k = 20) 
X_kbest = chi2_features.fit_transform(X, y) 
  
# Reduced features 
print('Original feature number:', X.shape[1]) 
print('Reduced feature number:', X_kbest.shape[1]) 
```

```python
chi = pd.DataFrame(X_kbest)
```

```python
chi.head()
```

```python
feature_names=list(X)
```

```python
mask = chi2_features.get_support() #list of booleans
new_features = [] # The list of your K best features

for bool, feature in zip(mask, feature_names):
    if bool:
        new_features.append(feature)
```

```python
new_features
dataframe = pd.DataFrame(X_kbest, columns=new_features)
```

```python
dataframe
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)
model.fit(dataframe)
```
